## Merge Data

```python
bj_states_gdf = states_gdf.merge(bjs_gdf, on="STUSPS", how="left")
bj_costco_states_gdf = bj_states_gdf.merge(costco_gdf, on="STUSPS", how="left")
bj_sc_costco_gdf = bj_costco_states_gdf.merge(sams_club_gdf, on="STUSPS", how="left")
```

```python
bj_sc_costco_gdf = bj_sc_costco_gdf.fillna(0)
```

```python
bj_sc_costco_gdf["BJS"] = bj_sc_costco_gdf["BJS"].astype(int)
bj_sc_costco_gdf["COSTCOS"] = bj_sc_costco_gdf["COSTCOS"].astype(int)
bj_sc_costco_gdf["SAMS_CLUBS"] = bj_sc_costco_gdf["SAMS_CLUBS"].astype(int)
```

```python
def more_stores(row) -> str:
    bj = row["BJS"]
    co = row["COSTCOS"]
    sc = row["SAMS_CLUBS"]
    if (bj > co) and (bj > sc):
        return "More BJ's Wholesales"
    if (co > bj) and (co > sc):
        return "More Costcos"
    if (sc > bj) and (sc > co):
        return "More Sam's Clubs"
    if bj == 0 and co == 0 and sc == 0:
        return "No Locations"
    if bj == sc:
        return "Same Amount of BJ's and Sam's Clubs"
```

```python
bj_sc_costco_gdf["label"] = bj_sc_costco_gdf.apply(lambda row: more_stores(row), axis=1)
```
